1) Nominal/OHE Encoding
2) Label & Ordinal Encoding
3) Target Guided Ordinal Encoding

# 1) Nominal/OHE Encoding

One hot encoding, also known as nominal encoding, is a technique used to represent categorical data as numerical data, which is more suitable for machine learning algorithms. In this technique, each category is represented as a binary vector where each bit corresponds to a unique category. For example, if we have a categorical variable "color" with three possible values (red, green, blue), we can represent it using one hot encoding as follows:

1. Red: [1, 0, 0]
2. Green: [0, 1, 0]
3. Blue: [0, 0, 1]

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# OneHotEncoder : Encode categorical features as a one-hot numeric array...
```

```python
## creating a simple dataframe..

df = pd.DataFrame({
    'color' : ['red', 'blue', 'green', 'green', 'red', 'blue']
})

df
```

```python
## creating an instance of OneHotEncoder...

encoder = OneHotEncoder()
```

```python
## perform fit and transform...

encoded = encoder.fit_transform(df[['color']]).toarray()

encoded
```

```python
## getting the feature names...

feature_names = encoder.get_feature_names_out(['color'])

feature_names
```

```python
encoder_df = pd.DataFrame(encoded,columns = feature_names)

encoder_df
```

```python
pd.concat([df,encoder_df],axis=1)
```

### Disadvantages:

1) High Dimensionality : If there are too many unique categories, it creates a large number of columns (called the curse of dimensionality). For instance, if you have 1,000 unique categories, you’ll end up with 1,000 new columns.

2) Sparse Matrix : Most of the values in the resulting dataset will be 0, which can increase memory usage and computation time.
